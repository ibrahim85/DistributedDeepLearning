## Multinomial Logistic Classification

```python
import numpy as np

def softmaxProbability(Y):
    '''
        Compute Softmax Probability

                W X + B = Y -> S 

    X matrix n(rows) by m(col)          
    each column of X is a data point with n features, 
    each column of S is a score vector with respect to the data 

    This function will turn Y, regression output, into S, softmax probability. 
    If Y consists only one raw, it means it has only 1 data point.

    '''
    # Vincent's solution is just one line
    # return np.exp(Y) / np.sum(np.exp(Y), axis = 0)
    
    if type(Y) is list or Y.ndim == 1:
        A = [np.exp(y) for y in Y]
        normal = sum(A)
        return A / normal
    
    Y_t = Y.T
    S = [] 
    for row in Y_t:
        A = [np.exp(r) for r in row]
        normal = sum(A)
        S.append(A/normal)
    
    return np.array(S).T
```

```python
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)]) 

import matplotlib.pyplot as plt  
%matplotlib inline
lineObjects = plt.plot(x, softmaxProbability(scores).T, linewidth=2)
plt.legend(iter(lineObjects), ('y_1', 'y_2', 'y_3'))
plt.ylabel('softmax')
plt.xlabel('x') 
plt.show() 
```

Above graph shows that for each data's regression output (x, 1, 0.2)  <br>
As x increases from -2 to 6, the probability of classifying it as y_1, y_2, y_3 changes.

## Cross Entropy Loss

```python
#  We want a lost function, so we can minimize something, which is the cross entropy

def oneHotEncoding(S):
    '''
             convert S to labels L
            
             WX + B = Y -> S -> L
             
    '''
    if type(S) is list or S.ndim == 1:
        m = np.max(S)
        A = [int(s == m)  for s in S] 
        return A 
    
    S_t = S.T
    L = [] 
    for row in S_t:
        m = np.max(row)
        L.append([int(s == m)  for s in row]) 
    
    return np.array(L).T
    
def crossEntropy(S, L = None):
    '''
        After convert S to labels L
            
             WX + B = Y -> S -> L
             
        cross entroy
            
            Loss = -SUM(L dot log(S))
            
        The label can also be given. That is the case of supervised training
    
    '''
    
    if L == None:
        if type(S) is list or S.ndim == 1:
            return - np.log(np.max(S))

        S_t = S.T

        return [- np.log(np.max(row)) for row in S_t]
    
    if type(S) is list or S.ndim == 1:
        return - np.dot(L, np.log(S))
    
    n, m = S.shape
    result = [0] * n
    L_t = L.T
    S_t = S.T
    
    for i in range(n):
        result[i] = - np.dot(L_t[i], S_t[i])
    return result 
```

```python
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)]) 
crossEntropy(softmaxProbability(scores))[-3:]
#clearly as x approaches 6, the loss decreases to 0
```

## Data Normalization

For good optimazation, we need good data and good labels
1. inital training uses certain labels (small variance). Leave tricky examples later
2. if possible, normalize data, so that each feature has mean 0 and equal variance split data into training, validation, test(never look at til very end)


## Rule of 30
If 30 more data get right labeled, the new model is statistically significantly better than the previous model. <br>This is important as we progress and adding more layers or more neurals. Typical size of validation set > 30000, so accuracy improving 0.1% is significant.

<a name = 'sgd'></a>
## Stochastic Gradient Descent

 we have

$$WX + B = Y \rightarrow S \rightarrow L$$

 we want to find $W, B$ such that $Loss()$ is minimal

$$arg_{W,B} \min\big (Loss(W, B)\big)$$

 Use SGD, because it scales well, 
 because it is faster and it works irrespective to the size and dimension of the data set
 because it does each dimension individually and during each step it takes a 
 fixed size of random sample of training data. 

To help SGD, we normalize data to have 0 mean, small equal variance for each feature. We initialize SGD, by randomly choosing 0 mean, small equal variance wights for each label.

SGD is done iteratively,
 
$$W_{new} = W_{old} - \text{StepSize} * \text{MultiDerivative} \big (Loss(W_{old})\big )$$
       
Multi-Derivative is the gradient. 

In reality, because each dimension is done separately, it tends to stumbling around zip-zap toward the optimal. The ramifications are

1. use Momentum to force it moving in the right directions. 
$$W_{new} = W_{old} - \text{StepSize} * (0.9 * \text{RunningAverageGradient} + \text{Gradient}\big (Loss(W_{old})\big)$$

In other words, past gradient are used with a 0.9 discount factor. 
There are other momentum formula: Nesterov Momentum, ...

2. step_size decay as iteration goes. Common ones are step decay, exponential decay, (1 / iteration) decay. But don't turn down the learning rate too soon, because it may hit the plateau. Also possible to apply different learning rates to each weight. 

3. picking normal distributed initial parameters. This is called batch normalization (ref: https://arxiv.org/abs/1502.0316 ). It will improve sgd significantly. 

4. decorrelated input features using PCA

5. For large dataset, use mini-batch instead of full-batch for speed and using mini-batch avoids check redundancy in the input data.

There are more. See: AdaGrad, RMSprop, and second order methods (quasi) Newton methods.

```python
'''
    what does the dataset look like?
    
'''


from scipy import ndimage
import numpy as np 

def montage(image_files):    
    image_size = 28 
    mm, nn = 15, 25
    M = np.zeros((mm * image_size, nn * image_size)) 
    
    k = 0
    for i in range(mm):
        for j in range(nn):
            slice_i, slice_j = i * image_size, j * image_size
            try:
                M[slice_i:slice_i + image_size, slice_j:slice_j + image_size] = \
                                        ndimage.imread(image_files[k]) 
                k += 1
            except IOError as e:  # some images are corrupted cannot read
                pass  
            
            if k >= nn*mm: 
                return M

import os
imagefiles = [os.path.join('notMNIST_small/B', i) for i in os.listdir('notMNIST_small/B') if '.png' in i]
plt.imshow(montage(imagefiles), cmap='Greys_r')
plt.title('B\'s in the test dataset')
plt.show() 
```

## Logistic Regression with Sciki-learn SGD  

If we had used usual logistic regression with batch gradient (sklearn.linear_model.LogisticRegression), it would be 100 time slower and same accuracy.

```python
'''
        input data from the disc, put them into appropriate shape
        
        only take 200000 out of 500000 training data for saving time.
        
'''

train_size = 200000
valid_size = 10000
test_size  = 10000


import pickle
pickle_file = 'notMNIST.p'
with open(pickle_file) as f:
        DataSetPickle = pickle.load(f)
        
train_data = DataSetPickle['train_data']
train_label = DataSetPickle['train_label']
valid_data = DataSetPickle['valid_data']
valid_label = DataSetPickle['valid_label']
test_data = DataSetPickle['test_data']
test_label = DataSetPickle['test_label']
del DataSetPickle

from sklearn.linear_model import LogisticRegression
import numpy as np

image_size = 28 

#Vincent's solution is just one line 
#dataset = dataset.reshape((-1, image_size * image_size)).astype(np.float32)

train_data_reshape = np.ndarray(shape=(train_size, image_size * image_size ), dtype=np.float32)
for i,image in enumerate(train_data):
    train_data_reshape[i] = image.reshape(1, image_size * image_size)
    
valid_data_reshape = np.ndarray(shape=(valid_size, image_size * image_size ), dtype=np.float32)
for i,image in enumerate(valid_data):
    valid_data_reshape[i] = image.reshape(1, image_size * image_size)
    
test_data_reshape = np.ndarray(shape=(test_size, image_size * image_size ), dtype=np.float32)
for i,image in enumerate(test_data):
    test_data_reshape[i] = image.reshape(1, image_size * image_size)
```

## Logistic Regression with SGD in TensorFlow

we do the same logistic regression with sgd in tensorFlow

```python
'''
        prepare for tensorFlow
        unlike scikit, the labels in TensorFlow need to be in 1-hot encode, not [0,1,2,3,4...]
'''
#modified from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/2_fullyconnected.ipynb

train_labels = (np.arange(10) == train_label[:,None]).astype(np.float32)
valid_labels = (np.arange(10) == valid_label[:,None]).astype(np.float32)
test_labels = (np.arange(10) == test_label[:,None]).astype(np.float32)

import tensorflow as tf

batch_size = 256 #fixed batch,  power of 2 good for parallelism
image_size = 28
num_labels = 10
num_steps = 2001

graph = tf.Graph()

with graph.as_default(): # setup the neural net

    # For parallel computing always need placeholder, 
    # batch_size is the fixed size of the SGD
    tf_train_dataset = tf.placeholder(tf.float32,
                                    shape=(batch_size, image_size * image_size))
    tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
    tf_valid_dataset = tf.constant(valid_data_reshape)
    tf_test_dataset = tf.constant(test_data_reshape)
  
    # Turning Parameters 
    # If we had put contant weights for the neurons, it would never learn.
    weights = tf.Variable(
               tf.truncated_normal([image_size * image_size, num_labels]))
    biases = tf.Variable(tf.zeros([num_labels]))
  
    # Formula
    #           XW + B -> Y -> S(Y)  => Loss(S(Y), labels) 
    
    logits = tf.matmul(tf_train_dataset, weights) + biases
    loss = tf.reduce_mean(
             tf.nn.softmax_cross_entropy_with_logits(logits, tf_train_labels))
  
    # Optimizer
    optimizer = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
  
    # Predictions for the training, validation, and test data.
    train_prediction = tf.nn.softmax(logits)
    valid_prediction = tf.nn.softmax(tf.matmul(tf_valid_dataset, weights) + biases)
    test_prediction = tf.nn.softmax(tf.matmul(tf_test_dataset, weights) + biases)
    
    
def accuracy(predictions, labels):
    correct_prediction = tf.equal(tf.argmax(predictions,1), tf.argmax(labels,1))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32)).eval()
      
```

```python
'''
    training begins
    
'''
#modified from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/2_fullyconnected.ipynb
import time

t0 = time.time()

with tf.Session(graph=graph) as session: # plug NN in 
    
    summary_writer = tf.train.SummaryWriter('log',session.graph)
    
    tf.initialize_all_variables().run()  # pick random weight  
    
    for step in range(num_steps): 
        
        # at every step only a fixed batch data are used
        # Depend on the num_steps and batch size, 
        # the same data in batch may train more than once.
        # people use epoch to mean the number of times
        # every data being seen
        
        offset = (step * batch_size) % (train_labels.shape[0] - batch_size) 
        batch_data = train_data_reshape[offset:(offset + batch_size), :]
        batch_labels = train_labels[offset:(offset + batch_size), :]  
        
        feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
        _, ls, predictions = session.run(
                      [optimizer, loss, train_prediction], feed_dict=feed_dict)
        
        if step % 300 == 0 or step == num_steps - 1:
            
            # mini-batch is a jargon for running SGD on small batch 
            print "MiniBatch loss at step %d: %f" % (step, ls)
            print "MiniBatch accuracy: %f" % accuracy(predictions, batch_labels)
            print "Validation accuracy: %f" % accuracy(
                        valid_prediction.eval(), valid_labels) 
        
    print "\nTest accuracy: %f" % accuracy(test_prediction.eval(), test_labels)

print 'running', (time.time() - t0), "seconds"
```

## Perceptron & Activation Functions

Let's recap, say one data point is fed into the NN

$$X = [...,...,...] \text{ with dimension } 28*28$$<br>
$$\text{regression output }Y\text{ is }[...,...,...]\text{ with dimension }10$$

We then compute softmax probability S(Y) and use 1-hot encoding to detemine its label (change soft classification to hard classificaton), and compute the loss:

$$X \times W + B \rightarrow Y \rightarrow S(Y)  \rightarrow Loss\big(S(Y), \text{labels}\big) $$
     
The regression assumes each feature is independent. In case some features are correlated, we can introduce non-linearity to the regression


$$(X \times W_1 + B_1) \rightarrow RELU(X \times W_1 + B_1) \rightarrow RELU(X \times W_1 + B_1) \times W_2 + B_2  \rightarrow Y$$  
     
An ReLU (rectifier linear unit) is just 

$$ReLU(a) = \max(a,0)$$ where a is number
   
And 

$$RELU(X \times W_1 + B_1)$$  is called Perceptron.

Because instead of fedding everything (linearly) into the regression, we now have the ability to turn off or remain on for some features. This new set of features may be more representative and are called first hidden layer neurons. 

Terminology: X - input layer, Y - output layer, others hidden layers

Caveat: Previously the shape of W must match X and Y, but now we are freely choose the number of hidden neurons. Say we want  

    hidden_nodes = 1024

That is 
    
    W_old = (28*28, 10)    
    
Now

    W_1.shape = (28*28, 1024) , W_2.shape = (1024, 10)

Other important activation functions: 

    Leaky ReLU  f(a) = a > 0 ? a: a*b  for some constant b 
    Maxout      Maxout = Max(W_1 * X + B_1, W_2 * X + B_2)
    
and some not so good    
    
    Sigmoid     sigma(a) = 1 / (1 + exp( -a ))
    Tanh        tanh(a) = 2 * sigma( 2 a ) -1

```python
'''
    set-up the simplest NN, with 1-hidden layer
    
'''

batch_size = 128
image_size = 28
num_labels = 10
num_steps = 2001
hidden_nodes = 1024

graph = tf.Graph()

with graph.as_default(): # 1-hidden layer, 1-input layer, 1-output layer
 
    with tf.name_scope('input_layer'):
        tf_train_dataset = tf.placeholder(tf.float32,
                         shape=(batch_size, image_size * image_size), name='train-data')
        tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels), name='train-labels')
        tf_valid_dataset = tf.constant(valid_data_reshape, name='valid-data')
        tf_test_dataset = tf.constant(test_data_reshape, name='test-data')
    
    with tf.name_scope('input_image'):
        image_shaped_input = tf.reshape(tf_train_dataset, [-1, 28, 28, 1])
        tf.image_summary('input-image', image_shaped_input, 10)
    
    def variable_summaries(var, name): 
 
        with tf.name_scope('summaries'):
            mean = tf.reduce_mean(var)
            tf.scalar_summary('mean/' + name, mean)
            
            with tf.name_scope('stddev'):
                stddev = tf.sqrt(tf.reduce_mean(tf.square(var - mean)))

            tf.scalar_summary('stddev/' + name, stddev)
            tf.scalar_summary('max/' + name, tf.reduce_max(var))
            tf.scalar_summary('min/' + name, tf.reduce_min(var))
            tf.histogram_summary(name, var)
    
    with tf.name_scope('weights'):
        weights_1 = tf.Variable(
               tf.truncated_normal([image_size * image_size, hidden_nodes]))
        variable_summaries(weights_1, 'hidden_layer/weights') 
        
        weights_2 = tf.Variable(
               tf.truncated_normal([hidden_nodes, num_labels])) 
        variable_summaries(weights_2, 'output_layer/weights') 
        
    with tf.name_scope('biases'):
        biases_1 = tf.Variable(tf.zeros([hidden_nodes]))  
        variable_summaries(weights_1, 'hidden_layer/biases') 

        biases_2 = tf.Variable(tf.zeros([num_labels])) 
        variable_summaries(weights_2, 'output_layer/biases') 
  
    # Formula
    #           ReLu(X * W_1  + B_1) * W_2 + B_2 -> Y -> S(Y)  => Loss(S(Y), labels) 
    
    with tf.name_scope('hidden_layer'):
        
        hidden_layer  = tf.nn.relu(tf.matmul(tf_train_dataset, weights_1) + biases_1) 
        tf.histogram_summary('hidden_layer', hidden_layer)
    
    with tf.name_scope('output_layer'):    
        logits = tf.matmul(hidden_layer, weights_2) + biases_2
        tf.histogram_summary('output_layer', logits)
        
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, tf_train_labels))
    
    tf.scalar_summary('loss', loss) 
    
    optimizer = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
   
    train_prediction = tf.nn.softmax(logits)
    
    valid_hidden_layer  = tf.nn.relu(tf.matmul(tf_valid_dataset, weights_1) + biases_1) 
    valid_logits = tf.matmul(valid_hidden_layer, weights_2) + biases_2
    valid_prediction = tf.nn.softmax(valid_logits)
    
    test_hidden_layer  = tf.nn.relu(tf.matmul(tf_test_dataset, weights_1) + biases_1) 
    test_logits = tf.matmul(test_hidden_layer, weights_2) + biases_2
    test_prediction = tf.nn.softmax(test_logits) 
    
    with tf.name_scope('accuracy'): 
        correct_prediction = tf.equal(tf.argmax(train_prediction,1), tf.argmax(tf_train_labels,1))
        accuracy_sum = tf.reduce_mean(tf.cast(correct_prediction, tf.float32)) 
        tf.scalar_summary('accuracy', accuracy_sum)
        
    merged = tf.merge_all_summaries()

def accuracy(predictions, labels):
    correct_prediction = tf.equal(tf.argmax(predictions,1), tf.argmax(labels,1))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32)).eval()    
```

```python
'''
    training begins
    
'''

import time

t0 = time.time()

with tf.Session(graph=graph) as session: 
    
    tf.initialize_all_variables().run()  
    
    train_writer = tf.train.SummaryWriter('log/train', session.graph) 
    
    for step in range(num_steps): 
         
        offset = (step * batch_size) % (train_label.shape[0] - batch_size) 
        batch_data = train_data_reshape[offset:(offset + batch_size), :]
        batch_labels = train_labels[offset:(offset + batch_size), :]   
        
        feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
        
        
        if step % 400 == 0 or step == num_steps - 1:
            
            _, ls, predictions, summary = session.run(
                      [optimizer, loss, train_prediction, merged], feed_dict=feed_dict)
            print "MiniBatch loss at step %d: %f" % (step, ls)
            print "MiniBatch accuracy: %f" % accuracy(predictions, batch_labels)
            print "Validation accuracy: %f" % accuracy(valid_prediction.eval(), valid_labels) 
            train_writer.add_summary(summary, step)
            train_writer.flush()
        else: 
             session.run([optimizer, loss], feed_dict=feed_dict)
            
    print "\nTest accuracy: %f" % accuracy(test_prediction.eval(), test_labels)

print 'running', (time.time() - t0), "seconds"
```

## L2 Regulation, DropOut, Learning Rate Decay 

We have shown above that adding addtional layers or changing number hidden neurons is a piece of cake in TensorFlow.

Next we discuss L2 regulation and drop out regulation. Because the number of parameters can be arbitrary large, it is possible that for a over-sized NN to have 100% accuracy on the training data and yet does poorly on the test data.

To fight that:

1. we add L2 norm |W| to the optimizer. So over-sized NN has very small |W| 
2. during every step of training, we randomly drop half of the hidden nodes (usually from the last highest hidden layer), so that the outcome will not be TOO dependent on some hidden features.
3. Max norm constraints. Say putting a cap on SDG, |W| = max(|W|,M) 

All of them solve the overfitting problem by dismantling any artificial dominant features.


```python
'''
    Applying L2 Regulation, DropOut, Learning Rate Decay and use 2-hidden layers, go deep

'''

batch_size = 128
image_size = 28
num_labels = 10
num_steps = 1001
hidden_nodes_1 = 2000
hidden_nodes_2 = 1000 
L2_Beta = 0.01 #L2 regulation, set to 0 if you don't want it 

'''
according to
http://yaroslavvb.blogspot.com/2011/09/notmnist-dataset.html?showComment=1458776573519#c1260767559237460571

    97% was achievable with 

    Test accuracy: 97.4%
    Validation accuracy: 91.9%
    Minibatch accuracy: 97.9%

    Parameters:
         
    Batch size: 1024
    Hidden layer 1 size: 4096
    Hidden layer 2 size: 2048
    Hidden layer 3 size: 1024
    Initial learning rate: 0.1
    Dropout probability: 0.5
    iteration 6000

'''

graph = tf.Graph()

with graph.as_default():#2-hidden layers, with L2 regulation, with dropout, with learning decay
 
    tf_train_dataset = tf.placeholder(tf.float32,
                                    shape=(batch_size, image_size * image_size))
    tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
    tf_valid_dataset = tf.constant(valid_data_reshape)
    tf_test_dataset = tf.constant(test_data_reshape)
   
    weights_1 = tf.Variable(
               tf.truncated_normal([image_size * image_size, hidden_nodes_1]))
    biases_1 = tf.Variable(tf.zeros([hidden_nodes_1]))
    
    weights_2 = tf.Variable(
               tf.truncated_normal([hidden_nodes_1, hidden_nodes_2]))
    biases_2 = tf.Variable(tf.zeros([hidden_nodes_2]))
    
    weights_3 = tf.Variable(
               tf.truncated_normal([hidden_nodes_2, num_labels]))
    biases_3 = tf.Variable(tf.zeros([num_labels]))
    
    hidden_layer_1  = tf.tanh(tf.matmul(tf_train_dataset, weights_1) + biases_1) 
    hidden_layer_2  = tf.tanh(tf.matmul(hidden_layer_1, weights_2) + biases_2)  
    #other activations: tf.sigmoid
    #                   tf.tanh
    #                   tf.nn.elu
    
    
    # drop half of nodes and the dropout function will rescale the undropped nodes
    hidden_layer_drop = tf.nn.dropout(hidden_layer_2, 0.5) # set to 1 if you do not want to drop
    
    logits = tf.matmul(hidden_layer_drop, weights_3) + biases_3
    
    # which is almost tf.nn.log_poisson_loss
    loss = (tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, tf_train_labels)) + 
                 L2_Beta * (tf.nn.l2_loss(weights_1) + tf.nn.l2_loss(weights_2) +
                               tf.nn.l2_loss(weights_3) +
                               tf.nn.l2_loss(biases_1) + tf.nn.l2_loss(biases_3))+
                               tf.nn.l2_loss(biases_2)  )
    
    #learning_rate = 0.5
    global_step = tf.Variable(0, trainable=False)
    starter_learning_rate = 0.1
    learning_rate = tf.train.exponential_decay(starter_learning_rate, global_step,
                                           100000, 0.96, staircase = True)
    #let learning rate decay exponentially
    optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss,
                                                                global_step=global_step)
                #other good ones: tf.train.MomentumOptimizer
                #                tf.train.AdagradOptimizer
                #                tf.train.ExponentialMovingAverage 
                #

    train_prediction = tf.nn.softmax(logits)
    
    valid_hidden_layer_1  = tf.tanh(tf.matmul(tf_valid_dataset, weights_1) + biases_1) 
    valid_hidden_layer_2  = tf.tanh(tf.matmul(valid_hidden_layer_1, weights_2) + biases_2)  
    valid_logits = tf.matmul(valid_hidden_layer_2, weights_3) + biases_3
    valid_prediction = tf.nn.softmax(valid_logits)
    
    
    test_hidden_layer_1  = tf.tanh(tf.matmul(tf_test_dataset, weights_1) + biases_1) 
    test_hidden_layer_2  = tf.tanh(tf.matmul(test_hidden_layer_1, weights_2) + biases_2)  
    test_logits = tf.matmul(test_hidden_layer_2, weights_3) + biases_3
    test_prediction = tf.nn.softmax(test_logits) 
    

def accuracy(predictions, labels):
    correct_prediction = tf.equal(tf.argmax(predictions,1), tf.argmax(labels,1))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32)).eval()
```

```python
'''
        training begins

'''

import time

t0 = time.time()

with tf.Session(graph=graph) as session: 
    
    tf.initialize_all_variables().run()  
    
    for step in range(num_steps): 
         
        offset = (step * batch_size) % (train_labels.shape[0] - batch_size) 
        batch_data = train_data_reshape[offset:(offset + batch_size), :]
        batch_labels = train_labels[offset:(offset + batch_size), :]  
        
        feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
        _, ls, predictions = session.run(
                      [optimizer, loss, train_prediction], feed_dict=feed_dict)
        
        if step % 300 == 0 or step == num_steps - 1:
            print "MiniBatch loss at step %d: %f" % (step, ls)
            print "MiniBatch accuracy: %f" % accuracy(predictions, batch_labels)
            print "Validation accuracy: %f" % accuracy(
                        valid_prediction.eval(), valid_labels)
    print "\nTest accuracy: %f" % accuracy(test_prediction.eval(), test_labels)

print 'running', (time.time() - t0) , "seconds"
```

# Convolution Neural Network for Image Recognition

For image classification, we know intensity, position, size don't matter. Thus instead of leaving these to the NN to learn, we should make them into the structure of NN such that it is roughly translational and scalar invariant. 

How to achieve these?

First we need a measure of similarity. Think each pixel as a feature. If two pictures are similar, their pixels distribution (the ratio of each component) are roughly the same, hence in the feature space, the two vectors are roughly parallel, not neccessary equal in lengh, because that depends on the intensity. This suggests we can make the cosine angle of the two vectors as a measure of similarity. For normalized pictures, dot products are sufficient.

What about translational invariance?

It is same task as for separating foreground and background. Say we are given an image of a car at the center and it is labeled as a car. For ML to learn, it will also pick a standard car, only has a car in it, a car patch, to map to the image and see how much they agree.

The process is as follows. 

We pick patch_size and number_of_patches (or use hyperparameterization), the ML will optimize and find the right patchs, then it will scan the whole image to find the car by computing the dot product of the patches and the image of moving region of the size of the patches. If the image has certain depth (say color in rgb instead of hex), then the patch should have the same depth, and the dot product will blend them together into one number.

How does the patch move? 

It is controlled by strikes and paddings. strike = [1,1] means moving in x, y directions 1 pixel at a time. Padding = 'VALID' means it never moves off the edge. Padding = 'SAME' means moving off edge and the center of the patch will collide with the edge. Values outside are set to 0.

    number_of_patches gives out-channels
    
Because each patch will filter the image and produce one layer of feature map, if we choose number_of_patches = 10, it will produce 10 feature maps, hence the depth of the convolution output is a image of depth 10.

<a name='worked'></a>
## Why does this work?

Below is a testing case, not trainning case. ML sees that when a car patch filter is applied the signal is strongest. The number of convolution is needed depends on the number of classification labels we need to train. The more labels, the more layers, because each layer will categorize the object a little bit more. In between, relu layers are also applied. They are there to turn off insignificant features. The pooling layers will be explained in the next section.

Let's summarize the genius of Yann LeCun's convolution networks:

   1. It asks ML to come up with patches: cars, cats,... to scan through the pictures, When the match is strong, the output signal will be strong. Thus the scanning process is almost translational invariant.
   2. It stores the convolution output as depth, called out-channels. Therefore in the subsequent convolution operations, the order of the out-channels are independent. Because in the dot product, switch components (x_1, x_2) and (y_1,y_2) makes no difference. This means for ML, applying car patch and applying cat patch are rougly independent.
   3. After the convolution process done, then a fully connected layer is applied. This layer will reshape all the out-channels and apply regression to them, thus it is called a fully connected layer.

```python
'''
    set-up convolution NN. Unlike general fully connected NN, 
    convNet data are 3-dimensional (x,y,z). z is the depth or out-channels 
    
'''

# modified from github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/4_convolutions.ipynb

import pickle
pickle_file = 'notMNIST.p'
with open(pickle_file) as f:
        DataSetPickle = pickle.load(f)
        
train_dataset = DataSetPickle['train_data']
train_labels = DataSetPickle['train_label']
valid_dataset = DataSetPickle['valid_data']
valid_labels = DataSetPickle['valid_label']
test_dataset = DataSetPickle['test_data']
test_labels = DataSetPickle['test_label']
del DataSetPickle

image_size = 28
num_labels = 10 

import numpy as np

def reformat(dataset, labels):
    ''' convNet dataset (not the label) are 3-dimensional '''
    dataset = dataset.reshape((-1, image_size, image_size, 1)).astype(np.float32)
    labels = (np.arange(num_labels) == labels[:,None]).astype(np.float32)
    return dataset, labels

train_dataset, train_labels = reformat(train_dataset, train_labels)
valid_dataset, valid_labels = reformat(valid_dataset, valid_labels)
test_dataset, test_labels = reformat(test_dataset, test_labels)

# modified from github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/4_convolutions.ipynb

batch_size = 24  # this is much small than before, due to my limited computing power and impatience
patch_size = 7   # so the patch is 7 * 7
depth = 5       # makes 5 feature maps
num_hidden = 64  # 1-hidden layer, call fully connected layer
num_steps = 801 # iterations

import tensorflow as tf
graph = tf.Graph()

with graph.as_default(): # 2-convolutional-layer and 1-hidden fully connected layer

    # Same as before except tf_train_dataset has 1 more dimension, 
    # it is required by nn.conv2d
    
    tf_train_dataset = tf.placeholder(tf.float32, shape=(batch_size, image_size, image_size, 1))
    tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
    tf_valid_dataset = tf.constant(valid_dataset)
    tf_test_dataset = tf.constant(test_dataset)
    
    #  
    #  2 layers of convolution
    #  filter.shape =  ([filter_height, filter_width, in_channels, out_channels])
    #
    layer1_weights = tf.Variable(tf.truncated_normal(
                    [patch_size, patch_size, 1, depth], stddev=0.1))
    layer1_biases = tf.Variable(tf.zeros([depth]))
    
    layer2_weights = tf.Variable(tf.truncated_normal(
                    [patch_size, patch_size, depth, depth], stddev=0.1))
    layer2_biases = tf.Variable(tf.constant(1.0, shape=[depth]))
    
    #
    # Optional hidden layer with 64 hidden neurals, this allows to use dropout regulation
    #
    # we could just put 
    #   layer3_weights = tf.Variable(tf.truncated_normal(
    #                [image_size / 4 * image_size / 4 * depth, num_labels], stddev=0.1))
    #   layer3_biases = tf.Variable(tf.constant(1.0, shape=[num_labels]))
    # and make it be the output final layer.
    #
    layer3_weights = tf.Variable(tf.truncated_normal(
                    [image_size / 4 * image_size / 4 * depth, num_hidden], stddev=0.1))
    layer3_biases = tf.Variable(tf.constant(1.0, shape=[num_hidden]))
    
    #
    # output final layer (also fully connected), convert hidden layer to the final output 10-labels
    #
    layer4_weights = tf.Variable(tf.truncated_normal([num_hidden, num_labels], stddev=0.1))
    layer4_biases = tf.Variable(tf.constant(1.0, shape=[]))

    def model(data):
        
        # 
        #    tf.nn.conv2d(input, filter, strides, padding)
        #    its output is 16 depth image with num pixel / 2 because of strike 2 used
        #
        conv = tf.nn.conv2d(data, layer1_weights, [1, 2, 2, 1], padding='SAME')
        # adding layer1_biases allows features to turn on/off
        hidden = tf.nn.relu(conv + layer1_biases) 
        
        # 
        #    its output is 16 depth image with num pixel / 4 because of another strike 2 used
        #    that is why layer3_weights.shape is image_size / 4 * image_size / 4 
        #
        conv = tf.nn.conv2d(hidden, layer2_weights, [1, 2, 2, 1], padding='SAME')
        hidden = tf.nn.relu(conv + layer2_biases)

        shape = hidden.get_shape().as_list()
        reshape = tf.reshape(hidden, [shape[0], shape[1] * shape[2] * shape[3]])
        hidden = tf.nn.relu(tf.matmul(reshape, layer3_weights) + layer3_biases)
    
        return tf.matmul(hidden, layer4_weights) + layer4_biases

    # Training 
    logits = model(tf_train_dataset)
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, tf_train_labels))

    # Optimizer
    optimizer = tf.train.GradientDescentOptimizer(0.05).minimize(loss)

    # Predictions 
    train_prediction = tf.nn.softmax(logits)
    valid_prediction = tf.nn.softmax(model(tf_valid_dataset))
    test_prediction = tf.nn.softmax(model(tf_test_dataset))
    
def accuracy(predictions, labels):
    correct_prediction = tf.equal(tf.argmax(predictions,1), tf.argmax(labels,1))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32)).eval()
```

```python
'''
    training begins

'''

# modified from github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/4_convolutions.ipynb

import time
import os
t0 = time.time()

with tf.Session(graph=graph) as session:
    
    tf.initialize_all_variables().run() 
    
    for step in range(num_steps):
        
        offset = (step * batch_size) % (train_labels.shape[0] - batch_size)
        batch_data = train_dataset[offset:(offset + batch_size), :, :, :]
        batch_labels = train_labels[offset:(offset + batch_size), :]
        
        feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
        _, ls, predictions = session.run(
          [optimizer, loss, train_prediction], feed_dict=feed_dict)
        
        if step % 200 == 0 or step == num_steps - 1:
            print('Minibatch loss at step %d: %f' % (step, ls))
            print('Minibatch accuracy: %f' % accuracy(predictions, batch_labels))
            print('Validation accuracy: %f' % accuracy(valid_prediction.eval(), valid_labels))
            
    print('\nTest accuracy: %f' % accuracy(test_prediction.eval(), test_labels))
    
    saver = tf.train.Saver()
    try:
        os.stat('backup')
    except:
        os.mkdir('backup')  
    saver.save(session, 'backup/convNet.ckpt')
    
print 'running', (time.time() - t0) , "seconds"
```

## Mix Pooling, 1x1 Convolution

Previously, before using convNet, to reach 90% accuracy, we had to run at least 30 seconds, now with convNet it takes 15 seconds. Can we do better?

Pooling

In the last example, we used strike 2 to reduce the image because strike 1 is too expensive to train (because strike 1 will produce 28 * 28 same size each time). But strike 2 is not very precise, pooling solves the problem. We can still do strike 1 and use pooling to reduce image size. It picks out the dominant features.

    Max Pooling(x) = max( neighbor of x )
    AVG Pooling(x) = avg( neighbor of x )
    
Famous pooling example: LeNet, AlexNet

1x1 Convolution

It is clear that the smaller the patch is, the faster it runs. But what if we set patch_size to 1, then the whole repeating scanning becomes just one matrix multiplication, which is not a deep network, but better than just logistic regression. 

```python
'''
    Apply Mix Pooling
    
'''

batch_size = 16 
patch_size = 7   #  set to 1 if you want 1x1 convolution
depth = 10    
num_hidden = 64  
num_steps = 601  

import tensorflow as tf
graph = tf.Graph()

with graph.as_default():  
 
    tf_train_dataset = tf.placeholder(tf.float32, shape=(batch_size, image_size, image_size, 1))
    tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size, num_labels))
    tf_valid_dataset = tf.constant(valid_dataset)
    tf_test_dataset = tf.constant(test_dataset)
     
    layer1_weights = tf.Variable(tf.truncated_normal(
                    [patch_size, patch_size, 1, depth], stddev=0.1))
    layer1_biases = tf.Variable(tf.zeros([depth]))
    
    layer2_weights = tf.Variable(tf.truncated_normal(
                    [patch_size, patch_size, depth, depth], stddev=0.1))
    layer2_biases = tf.Variable(tf.constant(1.0, shape=[depth]))
     
    layer3_weights = tf.Variable(tf.truncated_normal(
                    [image_size / 4  * image_size / 4 * depth, num_hidden], stddev=0.1))
    layer3_biases = tf.Variable(tf.constant(1.0, shape=[num_hidden]))
     
    layer4_weights = tf.Variable(tf.truncated_normal([num_hidden, num_labels], stddev=0.1))
    layer4_biases = tf.Variable(tf.constant(1.0, shape=[]))

    def model(data):
         
        conv = tf.nn.conv2d(data, layer1_weights, [1, 1, 1, 1], padding='SAME') 
        
        #
        #  pool here
        #
        pool = tf.nn.max_pool(conv, ksize = [1,2,2,1], strides = [1,2,2,1], padding='SAME')                     
        hidden = tf.nn.relu(pool + layer1_biases)  
        conv = tf.nn.conv2d(hidden, layer2_weights, [1, 1, 1, 1], padding='SAME')
        
        #
        #  pool here
        #
        pool = tf.nn.max_pool(conv, ksize = [1,2,2,1], strides = [1,2,2,1], padding='SAME') 
        hidden = tf.nn.relu(pool + layer2_biases)

        shape = hidden.get_shape().as_list()
        reshape = tf.reshape(hidden, [shape[0], shape[1] * shape[2] * shape[3]])
        hidden = tf.nn.relu(tf.matmul(reshape, layer3_weights) + layer3_biases)
    
        return tf.matmul(hidden, layer4_weights) + layer4_biases
 
    logits = model(tf_train_dataset)
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, tf_train_labels))
 
    optimizer = tf.train.GradientDescentOptimizer(0.05).minimize(loss)
 
    train_prediction = tf.nn.softmax(logits)
    valid_prediction = tf.nn.softmax(model(tf_valid_dataset))
    test_prediction = tf.nn.softmax(model(tf_test_dataset))
    
def accuracy(predictions, labels):
    correct_prediction = tf.equal(tf.argmax(predictions,1), tf.argmax(labels,1))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32)).eval()
```

```python
'''
    training begins

'''

import time

t0 = time.time()

with tf.Session(graph=graph) as session:
    
    tf.initialize_all_variables().run() 
    
    for step in range(num_steps):
        
        offset = (step * batch_size) % (train_labels.shape[0] - batch_size)
        batch_data = train_dataset[offset:(offset + batch_size), :, :, :]
        batch_labels = train_labels[offset:(offset + batch_size), :]
        
        feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels}
        _, ls, predictions = session.run(
          [optimizer, loss, train_prediction], feed_dict=feed_dict)
        
        if step % 200 == 0 or step == num_steps - 1:
            print('Minibatch loss at step %d: %f' % (step, ls))
            print('Minibatch accuracy: %f' % accuracy(predictions, batch_labels))
            print('Validation accuracy: %f' % accuracy(valid_prediction.eval(), valid_labels))
            
    print('\nTest accuracy: %f' % accuracy(test_prediction.eval(), test_labels))
    
print 'running', (time.time() - t0) , "seconds" 
```

```python
# modified from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/models/image/imagenet/classify_image.py

from IPython.display import display, Image
url = 'https://s-media-cache-ak0.pinimg.com/564x/74/36/5c/74365c7993daaf2502f7c3178a230ff9.jpg'
imagefile = 'inception_v3/test_image.jpg'    #works for jpg of any sizes
import urllib
urllib.urlretrieve(url, imagefile)
display(Image(imagefile)) 


num_top_predictions = 10   # number of softmax predictions

import re
class NodeLookup(object):
    
    """Converts integer node ID's to human readable labels."""

    def __init__(self):
        
        label_lookup_path = 'inception_v3/imagenet_2012_challenge_label_map_proto.pbtxt'
        uid_lookup_path = 'inception_v3/imagenet_synset_to_human_label_map.txt'
        self.node_lookup = self.load(label_lookup_path, uid_lookup_path)

    def load(self, label_lookup_path, uid_lookup_path):  
        '''
            load may take a few seconds
        '''
        proto_as_ascii_lines = tf.gfile.GFile(uid_lookup_path).readlines()
        uid_to_human = {}
        p = re.compile(r'[n\d]*[ \S,]*')
        for line in proto_as_ascii_lines:
            parsed_items = p.findall(line)
            uid = parsed_items[0]
            human_string = parsed_items[2]
            uid_to_human[uid] = human_string
 
        node_id_to_uid = {}
        proto_as_ascii = tf.gfile.GFile(label_lookup_path).readlines()
        for line in proto_as_ascii:
            if line.startswith('  target_class:'):
                target_class = int(line.split(': ')[1])
            if line.startswith('  target_class_string:'):
                target_class_string = line.split(': ')[1]
                node_id_to_uid[target_class] = target_class_string[1:-2]
 
        node_id_to_name = {}
        for key, val in node_id_to_uid.items():
            if val not in uid_to_human:
                tf.logging.fatal('Failed to locate: %s', val)
            name = uid_to_human[val].split(',')[0]
            node_id_to_name[key] = name

        return node_id_to_name

    def id_to_string(self, node_id):
        if node_id not in self.node_lookup:
            return ''
        return self.node_lookup[node_id]


image_data = tf.gfile.FastGFile(imagefile, 'rb').read()

with tf.gfile.FastGFile('inception_v3/classify_image_graph_def.pb', 'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())
    _ = tf.import_graph_def(graph_def, name='')

with tf.Session() as sess:
    # Some useful input, output, weights tensors:
    # 'softmax:0': A tensor containing the normalized prediction across
    #   1000 labels.
    # 'pool_3:0': A tensor containing the next-to-last layer containing 2048
    #   float description of the image.
    # 'DecodeJpeg/contents:0': A tensor containing a string providing JPEG
    #   encoding of the image.
    # Runs the softmax tensor by feeding the image_data as input to the graph.
    softmax_tensor = sess.graph.get_tensor_by_name('softmax:0')
    predictions = sess.run(softmax_tensor,
                           {'DecodeJpeg/contents:0': image_data})
    predictions = np.squeeze(predictions)

    node_lookup = NodeLookup()

    top_k = predictions.argsort()[-num_top_predictions:][::-1]
    for node_id in top_k:
        human_string = node_lookup.id_to_string(node_id)
        score = predictions[node_id]
        print('%s (probability = %.5f)' % (human_string, score))
```

## Word2Vec

Suppose we want to put every word into a vector of fixed length features, so that they can be fed to the NN. We can choose (there are many choices) to use the coordinates of the vectors as the order of frequence appearing in English language (i.e. in the wiki) 

    coordinates = ['the', 'or', 'and', 'one', 'in', 'a', 'to', 'zero', 'nine', 'two', ... ]

then we label each word as a vector, e.g.

    king = [..., 0.2, ...,0.014, ... , 0.082, ..., ...]

The vector gives the probabilities of king and some word from the corresponding coordinates to be appearing together in the text within some pre-defined window, e.g. within 2 words separation windows. 

If we take a dot product of two words, what does that imply?

    < king, queen > = #, 
    
It implies the probability of king and queen will be linked using one additional word (could be king,queen themselves).


What about 

    < king, king > ?
    
It should be 1 because king always linked to king with one additional word. This imposes a normalization constraint.

What about 

    king - man + woman ?
    
That turns out to a word very close to queen. We need to prove that multiplying above by any vector, V, we will get

    < (king - man + woman), V > ~ < queen, V >

That is because it is very likely that if V is equal distance to man and woman, it will be equal distance to king and queen. If V is only close to man, not woman, then it will be unlikely to close to queen.


These vector properties are very important, and later when we do dimension reduction, we would like to preserve those properties, hence we will be in the territory of Manifold reduction. 

In the reduced dimension world, the new coordinates will not be the real words, but it is not an issue, because we only care about relation between words, as long as we can still get the distance between two words using dot products and predict the next sequence of words using pluses / minuses. 



### skip-gram method for Word2Vec

If a word is very infrequent, using it as a component of the coordinate will result 0 for almost all words, so when we do dimension reduction, those rare dimensions will be squeezed into one dimemson. So we just count them as one word during word2vec.

```python
# import urllib
# url = 'http://mattmahoney.net/dc/text8.zip'
# urllib.urlretrieve(url, 'text8.zip')
```

```python
# modified from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/5_word2vec.ipynb
'''
    Create the coordinates
    
'''

import zipfile
import tensorflow as tf

words = []
with zipfile.ZipFile('text8.zip') as f:
    words = tf.compat.as_str(f.read(f.namelist()[0])).split()
print 'total words in the text:',len(words) 

import collections

vocabulary_size = 50000

count = [ ] 

# this line will get the frequency of all words in words, and 
# save the top 50000 to count and count is ordered by frequency
count.extend(collections.Counter(words).most_common(vocabulary_size - 1)) 

# word_vector rank words in freq

'''
      'UNK' in words => False
      so we just use UNK as a token for rara words
      
'''
dictionary = {'UNK':0}
for word, _ in count:
    dictionary[word] = len(dictionary)  
    
'''
    This is the coordinate of Word2Vec
'''
reverse_dictionary = dict(zip(dictionary.values(), dictionary.keys())) 

#
# data list, list frequency of each word in the order of word in the text
# This matches every word to the reverse_dictionary
# 
data = list() 
for word in words:
    if word in dictionary: 
        data.append(dictionary[word])
    else:  
        data.append(0) #uncommon words are set to 0 and matches 'UNK' token 

print '\nThis is the coordinate of Word2Vec:\n',
for i in range(1,11):
    print reverse_dictionary[i], '|',

last_word =  reverse_dictionary[50000-1]
c = 0
for [w,c] in count:
    if w == last_word: 
        break 
print '\n\nThe last component of the coordinate is \'', last_word,'\', and its freq is', c
```

```python
# modified from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/5_word2vec.ipynb

''' 
    A naive approach to build association between words within a window size m using loops
is at least (n^2)*(m^2), where n = 17005207

    A better way is to use softmax logistic regression with sgd in batches. Just pretend
the association as labels, and ask softmax to output the probability of labelings.

    Thus inputs will be like 
    
    ['a', 'human']
    ['a', 'house]
    ['a', 'hacker']
    ...
    
    Find all labels (within the window distance) for word 'a'. 
    The softmax will output the probability. Remember there will be 50000 labels and the text is
17005207 long. If we choose batch_size = 128 and num_strike = 2, then we will need at least

    17005207 / 128 * 2 = 265706 iterations to finish the whole text just for one epoch
    

    Set-up for logistic regression
    
'''

batch_size = 128
scan_window = 1 # window-distance, 2 means only words within 2 words distance.
num_strike = 2 #  cross product, which has to =< 2 * scan_window to be unique

# validation set to evaluate similarity on. 
valid_examples = [dictionary['life'],dictionary['war'],dictionary['history'],dictionary['day']]
valid_size = len(valid_examples)  

num_sampled = 64 # Number of sample for sample logistic loss. 
vocabulary_size = len(reverse_dictionary) # = 50000, size of the initial vector space
embedding_size = 50 # any number 
# 50 * vocabulary_size will be the total num of parameters for the softmax 

import numpy as np
import math
import random

data_index = 0
def generate_batch(batch_size, num_strike, scan_window): 
    global data_index # same as yield 
    
    assert batch_size % num_strike == 0   
    assert num_strike <= 2 * scan_window  
    
    batch = np.ndarray(shape=(batch_size), dtype=np.int32)
    labels = np.ndarray(shape=(batch_size, 1), dtype=np.int32)
    
    span = 2 * scan_window + 1  
    buffer = collections.deque(maxlen=span)
    
    for _ in range(span):
        buffer.append(data[data_index])
        data_index = (data_index + 1) % len(data)
        
    for i in range(batch_size / num_strike):
        target = scan_window  # target label at the center of the buffer
        targets_to_avoid = [ scan_window ]
        for j in range(num_strike):
            while target in targets_to_avoid:
                target = random.randint(0, span - 1)
            targets_to_avoid.append(target) 
            # batch return the constant center for the whole strike 
            # and labels are guaranteed to be not the same and from the same scan_window
            batch[i * num_strike + j] = buffer[scan_window]
            labels[i * num_strike + j, 0] = buffer[target]
            
        buffer.append(data[data_index]) 
        # after this is added, the first one will pop out
        # and uncommon words are added to the training as token 'UNK'
        
        data_index = (data_index + 1) % len(data)
    return batch, labels

'''

    As we can see below generate_batch function return association between each word in 
    the center of the scan_window. E.g. num_strike = 3, scan_window = 2.
    It produces 3 labels for each word, and the labels are within 2 distances away from the word.
    The outputs' type is int. Not words anymore.
    
'''
print('data:', [reverse_dictionary[di] for di in data[:8]])

for strike, scan_window in [(2, 1), (3, 2)]:
    data_index = 0
    batch, labels = generate_batch(batch_size=6, num_strike=strike, scan_window=scan_window)
    print('\nwith num_strike = %d and scan_window = %d:' % (strike, scan_window)) 
    print('batch:', [reverse_dictionary[bi] for bi in batch])
    print('labels:', [reverse_dictionary[li] for li in labels.reshape(6)])

graph = tf.Graph()

with graph.as_default():
    

    # Input data
    train_dataset = tf.placeholder(tf.int32, shape=[batch_size])
    train_labels = tf.placeholder(tf.int32, shape=[batch_size, 1])
    valid_dataset = tf.constant(valid_examples, dtype=tf.int32)

    # Variables
    embeddings = tf.Variable(tf.random_uniform([vocabulary_size, embedding_size], -1.0, 1.0))
    softmax_weights = tf.Variable(tf.truncated_normal([vocabulary_size, embedding_size],
                         stddev = 1.0 / math.sqrt(embedding_size)))
    softmax_biases = tf.Variable(tf.zeros([vocabulary_size]))

    # Model
    #
    # Look up embeddings for inputs
    # it is almost the same as matrix multiplication as logistic regression
    #
    #            W * X + B => Y => S(Y) => encode(S(Y)) => Min(Loss(encode(S(Y)), Labels))
    #
    # but here each X any one of the 50000 words, so we use embedding_lookup 
    # to convert it to dict keys. Then it will do the matrix multiplication.
    # The complexity of the hashed keys are controlled by embedding_size
    #   
    embed = tf.nn.embedding_lookup(embeddings, train_dataset)
    # 
    # the full 1-hot encoding is 50000 long, so compute the softmax loss is expensive, 
    # instead using a sample of the labels each time
    #
    loss = tf.reduce_mean(tf.nn.sampled_softmax_loss(
                softmax_weights, softmax_biases, embed,
                     train_labels, num_sampled, vocabulary_size))

    # Optimizer.
    # Note: The optimizer will optimize the softmax_weights AND the embeddings.
    # This is because the embeddings are defined as a variable quantity and the
    # optimizer's `minimize` method will by default modify all variable quantities 
    # that contribute to the tensor it is passed. (Lazy evaluation is triggered)
    optimizer = tf.train.AdagradOptimizer(1.0).minimize(loss)

    # Compute the similarity between minibatch examples and all embeddings.  
    norm = tf.sqrt(tf.reduce_sum(tf.square(embeddings), 1, keep_dims=True))
    normalized_embeddings = embeddings / norm
    valid_embeddings = tf.nn.embedding_lookup(normalized_embeddings, valid_dataset)
    similarity = tf.matmul(valid_embeddings, tf.transpose(normalized_embeddings))
```

## Manifold Reduction: tSNE & Kullback-Leibler Divergence
 
Suppose we have some high dimensional data ${X}$ and we would like to construct a probability model and the representation ${Y}$ in 2 or 3 dimensions. Recall entropy of information, which is a single number measuring for a given probability $P(X)$ over some probabilistic space $X$, how chaotic the distribution is. 

$$H = -\sum_{x\in X}{P(x)\log P(x)}$$
    
In the cell below I am showing that the constant distribution has the highest entropy. 

```python
import math
import random
N = 100
P = np.array([random.random() for a in range(N)])
P = P/sum(P)
print 'random distribution:', -sum(a*math.log(a,2) for a in P )
print 'constant distribution:', -math.log(1.0/N,2)
```

We used cross entropy to discuss cost function of softmax. So what is it really? Softmax cross entropy $P(X_i), Q(Y_i) $ are probabilities on the label space. But it works on general spaces, 

$$ -\sum_{i}{P(X_i)\log Q(Y_i)}$$ 

One can show that if the two distributions are identical, the cross entropy is at the lowest. So what is cross entropy? It measures the randomness in identifying $Y_i => X_i$

```python
import math
import random
import copy
N = 100
P = P
Q = np.array([random.random() for a in range(N)]) 
P = P/sum(P)
Q = Q/sum(Q) 
print 'two random distributions:', -sum(a*math.log(b,2) for a,b in zip(P,Q))
print 'two identical distributions:', -sum(a*math.log(a,2) for a in P)
print 'two constant distributions:', -math.log(1.0/N,2)
```

### Kullback-Leibler Divergence

Kullback–Leibler divergence is defined to be the difference of the two entropies above.

$$KL(X||Y) = \text{cross entropy} - H(X)= -\sum_{i}{P(X_i)\log Q(Y_i)} - \big [-\sum_{X}{P(X)\log P(X)}\big ] = \sum_{i}{P(X_i)\log \frac{P(X_i)}{Q(Y_i)}}$$

KL is non-nagetive and only zero when $P = Q$. Physically it means there were randomness in both $P,Q$ and assigning $Y_i => X_i$, the entropy of $P$ can only go up. Mathematically one can prove non-nagetivity of KL using Jensen's inequality. The proof is exactly the same as in proving why EM (expectation maximization) algorithm converges. 

The genius of Geoffrey Hinton, who invented tSNE and SNE is that he recoginzed the importance of KL, just like in transforming graph theory. When transforming a high dimensional representaion down to a very low dimension, it is less efficient to focus on minimizing cross entropy, which is inherited from the entropy of $P$, but to focus on minimizing KL.

The tSNE algorithm is on page 9 http://www.cs.toronto.edu/~hinton/absps/tsne.pdf. Basically it minimizes KL and uses the following to simplify derivatives for steepest descent

$$e^{-x} = 1 - x = \frac{1}{1+x}$$

for the low dimensional representation.

### Mutual Information
For completeness, let me add two other related concepts: conditional entropy and mutual information. They are too talking about two random variables with two probabilities, $P(X), Q(Y)$. Conditional entropy is defined to be randomness of the conditional probability of $X$ given $Y$, denote $J(x,y)$ be joint probability

$$H(X|Y)=-\sum_{y\in Y} Q(y) \sum_{x\in X} P(x|y)\log P(x|y) = -\sum_{x\in X,y\in Y} J(x,y) \log \frac{J(x,y)}{Q(y)} $$

And mutual information is the difference

$$I(X,Y) = H(X) - H(X|Y) = -\sum_{x\in X,y \in Y} J(x,y)\log P(x)-H(X|Y)= \sum_{x\in X,y\in Y} J(x,y) \log \frac{J(x,y)}{P(x)Q(y)}$$

the last line uses the identity that $\sum_{ y \in Y} J(x,y)=P(x)$. Mutual information is also non-negative and zero iff $X,Y$ are independent. 

Once we saw how Kullback-Leibler Divergence and mutual information were defined from, we would be able to better understand their usages. Both are important concepts in graph theory, e.g. construct and learn decision tree structure from probabilistic graph models.

## Long Short-Term Memory RNN with Dropout Regulation

The idea of recurrent neural networks (RNN) has been floating around for long time as a best candidate for dealing with time series, because it's conceptually related to Autoregression and structurally related to probabilistic graph model: Hidden Markov Models. However it was not till recently when large companies devoted to it and large dataset being used to train it, RNN is now able to reach an impressive accuracy on voice recognition, https://research.googleblog.com/2015/09/google-voice-search-faster-and-more.html and near human level accuracy on text translation (the article https://research.googleblog.com/2016/09/a-neural-network-for-machine.html emphasised its success in "notoriously difficult language pair" Chinese to English translation).

### What is RNN and LSTM?

It is a recursive (people call a generalized RNN a recursive NN) neural network. Each point in the time series has three inputs and two outputs.

- Three Inputs: 
    - The current value in the series $$i_t$$<br>
    - Last predicted value $$o_{t-1}$$<br>
    - The state of the last value, which is analogous to HMM hidden state, and it summarizes all history from the past. $$s_{t-1}$$<br>
- Two Outputs: 
    - The next value need to predict $$o_t$$<br>
    - The state of current value, which is a summary of history from its past to current $$s_t$$<br>
- LSTM Mechanism
    - It has Three hidden gates. Each gate is a hidden layer of the deep network connecting the three inputs and two outputs. The reason we cannot just connect two outputs to the three inputs using 2 1-layer perceptrons is that perceptron fails at differentiating pattern if pattern can wrap around. 2-layer NN solves that problem.<br><br>
    - Input Gate
    $$I_t = \sigma(X_I * i_{t} + M_I * o_{t-1} + B_I)$$<br>
    - Gorget Gate 
    $$F_t = \sigma(X_F * i_{t} + M_F * o_{t-1} + B_F)$$<br>
    - Output Gate 
    $$O_t = \sigma(X_O * i_{t} + M_O * o_{t-1} + B_O)$$<br>

    - Output State  
    $$s_t = F_t * s_{t-1} + I_t * \tanh(X_C * i_{t} + M_C * o_{t-1} + B_C)$$<br> 
    - Output Value   
    $$o_t =  O_t * \tanh(s_t)$$ <br>
    - Optimizer   
    $$ Loss =  mean \bigg ( \sum \log p_{target} \bigg ) $$ <br>
    $$ perplexity =  e^{Loss} $$ <br>  


where $X, M, B$ are the hidden weights and $\sigma$ is the sigmoid function. The loss function is the same as before, using the softmax cross entropy, take logarithm of the soft probability of the predicting label. We are optimizing the loss function and we use the perplexity to represent the error. Because we used hidden layer to manipulate input/output, it allows us to use arbitary number of hidden nodes ('num_nodes'), so except the input $i_t$ is in 1-hot encode i.e. vector of dimension of the vocabularry, all other four values (2 other inputs and 2 outputs) are vectors of the dimension of the hidden neurons, which can be any value.

```python
'''
    Load text from text8.zip and create training and validation sets
    
    
    I wish I could do the same word to word manipulation as before, 
    
    but my computer power and patiency limit it. 
    
    The follow program will not try to predicts sentences, 
    
    because that requires to run the time series on words, 
    
    and let RNN figure out grammar, meaning of words and word associations on its own. 
    
    
    
    Instead we ask the program to predict English words.  That is we feed RNN
    
    time series of characters (a-z including spaces), 
    
    the program doesn't know spaces will separate English words.
    
    Then we ask the program to predict the next character (a-z including spaces)
    
    so that the outputs will eventually all look like real words.
    
    
    
    Although the program does not do word to word manipulation, but it can 
    
    perform word to word manipulation with very little modification: 
    
    change vocabulary to 50000 and for such large vocabulary use embedding_look_at 
    
    and sampling like before word2vec, avoid naive matrix multiplication with 50000 rows.
    
    
    
    
    That is because here the number of choices ML has to make is the vocabulary = 26 + 1 
    
    (1 for the space). For word to word manipulation, the number of choices is vocabulary = 50000.
    
    so to train word-to-word, like the im2txt we discussed before, will take several weeks
    
    https://github.com/tensorflow/models/tree/master/im2txt#getting-started
    
        "
         In our experience on an NVIDIA Tesla K20m GPU the initial 
         training phase takes 1-2 weeks. The second training phase 
         may take several additional weeks to achieve peak performance 
         (but you can stop this phase early and still get reasonable 
         results). 
        "
     
'''

import zipfile
import tensorflow as tf

text = []
with zipfile.ZipFile('text8.zip') as f:
    # before we split into words, now all text
    # so it is just a continuous stream of wikipedia articles
    text = tf.compat.as_str(f.read(f.namelist()[0])) 
print 'total length in the text:',len(text) 

# save the first 1000 characters as validation set
# to check if ML can predict character by character
valid_size = 1000
valid_text = text[:valid_size]
train_text = text[valid_size:]
train_size = len(train_text) 
```

```python
# coding=utf-8
# copied from https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/udacity/6_lstm.ipynb


import numpy as np
import tensorflow as tf
import random
import string

batch_size = 64
num_unrollings = 10
vocabulary_size = 27
num_nodes = 64 # hidden neurons

'''

    prepare for RNN
     

'''

class BatchGenerator(object):
    '''
            We created this object, because we wanted to store the array of cursors
            
            When the 'next' method gets called, it will return batch_size = 64 of
            
            time series, each has (num_unrollings + 1 = 11) characters long. 
            
            The first character is the last character in the last batch of the corresponding
            
            time series, because in the last batch the last char was not used for training.
            
            The very first cursors (there are 64 of them) will be located at the beginning 
            
            of the 64 equally divided segments of the text. 
    
    '''
    def __init__(self, text, batch_size, num_unrollings):
        self._text = text
        self._text_size = len(text)
        self._batch_size = batch_size
        self._num_unrollings = num_unrollings
        segment = self._text_size / batch_size
        
        # initial cursor
        self._cursor = [ offset * segment for offset in range(batch_size)]
        
        # inital _last_batch 
        self._last_batch = self._next_batch()

    def _next_batch(self):
        """
            helper
            
            Generate a single batch from the current cursor position in the data.
        """
        batch = np.zeros(shape=(self._batch_size, vocabulary_size), dtype=np.float)
        
        for b in range(self._batch_size): # 64 loops
            
            #1-hot encoding
            batch[b, char2id(self._text[self._cursor[b]])] = 1.0
            
            # cursor moves one char for all 64 time series during the loop
            self._cursor[b] = (self._cursor[b] + 1) % self._text_size
        return batch
  
    def next(self):
        """
            Generate the next array of batches from the data. The array consists of
            
            the last batch of the previous array, followed by num_unrollings new ones.
        
        """
        batches = [self._last_batch]
        for step in range(self._num_unrollings):
              batches.append(self._next_batch())
        #
        # save the last 64 char for the next batch
        #
        self._last_batch = batches[-1]
        
        # 64 times series, each is 11-character long
        # e.g. batches[0] gives 64 characters encoded in 1-hot encoding
        #      batches[1] gives the next 64 characters in the time series
        return batches

    
def char2id(char):
    #
    #   turn char to id 
    #
    if char in string.ascii_lowercase:
        return ord(char) - ord('a') + 1 
    return 0
  
def id2char(dictid): 
    #
    #   turn id to char
    #
    if dictid > 0:
        return chr(dictid + ord('a') - 1) 
    return ' '

def characters(probabilities):
    #
    #   Turn array of 1-hot encoding into array of character representations.
    #
    return [id2char(c) for c in np.argmax(probabilities, 1)]
  

def logprob(predictions, labels):
    #
    #   Log-probability of the true labels in a predicted batch.
    #
    predictions[predictions < 1e-10] = 1e-10
    return np.sum(np.multiply(labels, -np.log(predictions))) / labels.shape[0]

def sample_distribution(distribution):
    #
    #    Sample one element from a distribution assumed to be an array of normalized   
    #    probabilities.
    #
    r = random.uniform(0, 1)
    s = 0
    for i in range(len(distribution)):
        s += distribution[i]
        if s >= r:
            return i
    return len(distribution) - 1

def sample(prediction):
    #
    # Turn a (column) prediction into 1-hot encoded samples.
    #
    p = np.zeros(shape=[1, vocabulary_size], dtype=np.float)
    p[0, sample_distribution(prediction[0])] = 1.0
    return p

def random_distribution():
    #
    #  Generate a random column of probabilities.
    #
    b = np.random.uniform(0.0, 1.0, size=[1, vocabulary_size])
    return b/np.sum(b, 1)[:,None]


train_batches = BatchGenerator(train_text, batch_size, num_unrollings)
valid_batches = BatchGenerator(valid_text, 1, 1)

graph = tf.Graph()
with graph.as_default():

    # Parameters
    # Input gate: input, previous output, and bias.
    ix = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    im = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    ib = tf.Variable(tf.zeros([1, num_nodes]))
    # Forget gate: input, previous output, and bias.
    fx = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    fm = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    fb = tf.Variable(tf.zeros([1, num_nodes]))
    # Memory cell: input, state and bias.                             
    cx = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    cm = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    cb = tf.Variable(tf.zeros([1, num_nodes]))
    # Output gate: input, previous output, and bias.
    ox = tf.Variable(tf.truncated_normal([vocabulary_size, num_nodes], -0.1, 0.1))
    om = tf.Variable(tf.truncated_normal([num_nodes, num_nodes], -0.1, 0.1))
    ob = tf.Variable(tf.zeros([1, num_nodes]))
    # Variables saving state across unrollings.
    saved_output = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
    saved_state = tf.Variable(tf.zeros([batch_size, num_nodes]), trainable=False)
    # Classifier weights and biases.
    w = tf.Variable(tf.truncated_normal([num_nodes, vocabulary_size], -0.1, 0.1))
    b = tf.Variable(tf.zeros([vocabulary_size]))

    # Define LSTM cell
    #
    # Here we have to hand make lstm_cell instead grab 
    # lstm_cell = tf.nn.rnn_cell.BasicLSTMCell and tf.nn.rnn_cell.MultiRNNCell() wrapper
    # because in training batch = 64, while in testing batch = 1
    # 
    def lstm_cell(i, o, state):
        """
            Create a LSTM cell. See e.g.: http://arxiv.org/pdf/1402.1128v1.pdf
            
            Note that in this formulation, we omit the various connections between the
            
            previous state and the gates.
        """
        input_gate = tf.sigmoid(tf.matmul(i, ix) + tf.matmul(o, im) + ib)
        forget_gate = tf.sigmoid(tf.matmul(i, fx) + tf.matmul(o, fm) + fb)
        output_gate = tf.sigmoid(tf.matmul(i, ox) + tf.matmul(o, om) + ob)
        
        update = tf.matmul(i, cx) + tf.matmul(o, cm) + cb
        state = forget_gate * state + input_gate * tf.tanh(update)
        return output_gate * tf.tanh(state), state

    # Input data
    train_data = list()
    for _ in range(num_unrollings + 1):
        train_data.append(tf.placeholder(tf.float32, shape=[batch_size,vocabulary_size]))
    #
    # inputs are the first 10 char in the time series, 
    # the 11th char is the label of the 10th char, thus not to be used for input
    #
    train_inputs = train_data[:num_unrollings]
    #
    # labels are inputs shifted forword by one time step
    #
    train_labels = train_data[1:]  

    # Unrolled LSTM loop.
    outputs = list()
    #
    # saved_output, saved_state from the previous character go into the next character
    #
    output = saved_output
    state = saved_state
    for i in train_inputs:
        output, state = lstm_cell(i, output, state)
        outputs.append(output)

    # 
    # saved_output, saved_state from the previous batch go into the next batch 
    #
    with tf.control_dependencies([saved_output.assign(output), saved_state.assign(state)]):
        # Classifier.
        logits = tf.nn.xw_plus_b(tf.concat(0, outputs), w, b)
        loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(
                logits, tf.concat(0, train_labels)))

    # Optimizer.
    global_step = tf.Variable(0)
    learning_rate = tf.train.exponential_decay(10.0, global_step, 5000, 0.1, staircase=True)
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    gradients, v = zip(*optimizer.compute_gradients(loss))
    gradients, _ = tf.clip_by_global_norm(gradients, 1.25)
    optimizer = optimizer.apply_gradients(zip(gradients, v), global_step=global_step)

    # Predictions
    train_prediction = tf.nn.softmax(logits)

    # Prediction eval: batch 1, 1 character at a time, 
    sample_input = tf.placeholder(tf.float32, shape=[1, vocabulary_size])
    saved_sample_output = tf.Variable(tf.zeros([1, num_nodes]))
    saved_sample_state = tf.Variable(tf.zeros([1, num_nodes]))
    reset_sample_state = tf.group(
            saved_sample_output.assign(tf.zeros([1, num_nodes])),
            saved_sample_state.assign(tf.zeros([1, num_nodes])))
    sample_output, sample_state = lstm_cell(sample_input, saved_sample_output, saved_sample_state)
    
    with tf.control_dependencies([saved_sample_output.assign(sample_output),
                            saved_sample_state.assign(sample_state)]):
        sample_prediction = tf.nn.softmax(tf.nn.xw_plus_b(sample_output, w, b))
```

# Application: Sequence-to-Sequence Models


In the example above I demonstrated using RNN to predict the next element in the sequence. But it will be the same if we have two sequences. One is the time series of input, and the other is the label. The label doesn't need to be the input time series itself, but a parallel time series. Thus we have a sequence-to-sequence model.
